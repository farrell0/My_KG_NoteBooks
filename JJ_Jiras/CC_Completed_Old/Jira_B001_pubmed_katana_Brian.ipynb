# `(Affiliation)<--(Author)-->(Article)-->(MeshTerm)`

```python
%%time
graph.query('''MATCH (mt:MeshTerm)
               WHERE toLower(mt.DescriptorName) CONTAINS "neoplasm"
               WITH mt
               MATCH (mt)<-[:HAS_MESHTERM]-(art:Article)
               MATCH (a:Author)-[:HAS_ARTICLE]->(art)
               WHERE art.PubYear >= 2012
               MATCH (a)-[haf:HAS_AFFILIATION]->(af:Affiliation)
               WHERE toLower(haf.Institution) CONTAINS "yale"
               OR toLower(haf.Institution) CONTAINS "harvard"
               OR toLower(haf.Institution) CONTAINS "hopkin"
               OR toLower(haf.Institution) CONTAINS "mit"
               OR toLower(haf.Institution) CONTAINS "stanford"
               RETURN a, art, af, mt
               LIMIT 1000''', contextualize=True).view()
```

```python
%%time
graph.query('''MATCH (mt:MeshTerm)
               MATCH (art:Article)-[:HAS_MESHTERM]->(mt)
               MATCH (a:Author)-[:HAS_ARTICLE]->(art)
               WHERE art.PubYear >= 2012
               MATCH (a)-[haf:HAS_AFFILIATION]->(:Affiliation)
               WHERE toLower(haf.Institution) CONTAINS "yale"
               AND toLower(mt.DescriptorName) CONTAINS "neoplasm"
               RETURN a.LastName, a.ForeName, a.pr AS PageRank, haf.Institution
               ORDER BY PageRank DESC
               LIMIT 1000''', contextualize=True).view()
```
